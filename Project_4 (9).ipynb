MODEL BUILDING

```python
# Features and Target
X = df[['Open','High','Low','Volume']]
y = df['Close']

print("Feature Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
# Time Series Split (80-20)

X = df[['Open','High','Low','Volume']]
y = df['Close']

split = int(len(df) * 0.8)

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]

print("Train Size:", X_train.shape)
print("Test Size:", X_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None]
}

grid = GridSearchCV(RandomForestRegressor(), param_grid, cv=3)
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)

best_rf = grid.best_estimator_
rf_tuned_pred = best_rf.predict(X_test)
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Close'][:split], label='Train')
plt.plot(df['Close'][split:], label='Test')
plt.legend()
plt.title("Train-Test Split")
plt.show()
```

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
```

XGBOOST

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(n_estimators=200, learning_rate=0.05)
xgb.fit(X_train, y_train)

xgb_pred = xgb.predict(X_test)
```

MODEL EVALUATION

MODEL COMPARISON TABLE

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model(name, y_true, y_pred):
    print(f"\n{name} Results")
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("RMSE:", np.sqrt(mean_squared_error(y_true, y_pred)))
    print("R2 Score:", r2_score(y_true, y_pred))
```

```python
evaluate_model("Random Forest Tuned", y_test, rf_tuned_pred)
```

```python
results = pd.DataFrame({
    "Model": ["Linear Regression","Random Forest Tuned","XGBoost"],
    "MAE": [
        mean_absolute_error(y_test, lr_pred),
        mean_absolute_error(y_test, rf_tuned_pred),
        mean_absolute_error(y_test, xgb_pred)
    ],
    "RMSE": [
        np.sqrt(mean_squared_error(y_test, lr_pred)),
        np.sqrt(mean_squared_error(y_test, rf_tuned_pred)),
        np.sqrt(mean_squared_error(y_test, xgb_pred))
    ],
    "R2 Score": [
        r2_score(y_test, lr_pred),
        r2_score(y_test, rf_tuned_pred),
        r2_score(y_test, xgb_pred)
    ]
})

results
```

LAG FEATURE MODEL

```python
df_lag = df.copy()

df_lag['Lag_1'] = df_lag['Close'].shift(1)
df_lag['Lag_2'] = df_lag['Close'].shift(2)
df_lag['Lag_3'] = df_lag['Close'].shift(3)

df_lag.dropna(inplace=True)
```

```python
X_lag = df_lag[['Lag_1','Lag_2','Lag_3']]
y_lag = df_lag['Close']

split = int(len(df_lag)*0.8)

X_train_lag = X_lag[:split]
X_test_lag = X_lag[split:]

y_train_lag = y_lag[:split]
y_test_lag = y_lag[split:]

rf_lag = RandomForestRegressor(n_estimators=200)
rf_lag.fit(X_train_lag, y_train_lag)

lag_pred = rf_lag.predict(X_test_lag)

evaluate_model("Lag Model RF", y_test_lag, lag_pred)
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[['Close']])

X_lstm = []
y_lstm = []

for i in range(60, len(scaled_data)):
    X_lstm.append(scaled_data[i-60:i])
    y_lstm.append(scaled_data[i])

X_lstm, y_lstm = np.array(X_lstm), np.array(y_lstm)

split_lstm = int(len(X_lstm)*0.8)

X_train_lstm = X_lstm[:split_lstm]
X_test_lstm = X_lstm[split_lstm:]

y_train_lstm = y_lstm[:split_lstm]
y_test_lstm = y_lstm[split_lstm:]

model = Sequential()
model.add(LSTM(50))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
model.fit(X_train_lstm, y_train_lstm, epochs=10, batch_size=32)

lstm_pred = model.predict(X_test_lstm)
```

    FUTURE 30 DAYS FORECAST

```python
future_days = 30
last_60_days = scaled_data[-60:]

future_predictions = []
current_input = last_60_days

for _ in range(future_days):
    current_input_reshaped = current_input.reshape(1,60,1)
    next_pred = model.predict(current_input_reshaped)
    
    future_predictions.append(next_pred[0][0])
    current_input = np.append(current_input[1:], next_pred, axis=0)

future_predictions = scaler.inverse_transform(
    np.array(future_predictions).reshape(-1,1)
)

plt.figure(figsize=(12,6))
plt.plot(future_predictions)
plt.title("Next 30 Days Future Prediction")
plt.show()
```
